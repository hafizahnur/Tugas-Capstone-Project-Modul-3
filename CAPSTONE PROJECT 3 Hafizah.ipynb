Selanjutnya, kita akan membangun model klasifikasi yang akan membantu perusahaan dalam memprediksi probabilitas seorang pelaku perjalanan perjalanan untuk memenuhi syarat atau tidak memenuhi syarat untuk menerima klaim.

##### **Evaluasi Metrik**

- **True Positif**: Calon pelaku perjalanan mengajukan klaim dan berhak menerima klaim, artinya prediksinya sesuai.

- **False Positif**: Calon pelaku perjalanan mengajukan klaim, tetapi model memprediksi bahwa calon pelaku perjalanan tidak berhak menerima klaim. Ini adalah tipe error 1.

- **False Negatif**: Calon pelaku perjalanan tidak mengajukan klaim, tetapi model memprediksi bahwa calon pelaku perjalanan berhak menerima klaim. Ini adalah tipe error 2.

- **True Negatif**: Calon perjalanan perjalanan tidak mengajukan klaim, dan model memprediksi dengan benar bahwa calon pelaku perjalanan tidak menerima klaim.


**Tipe error 1: False Positif**

Dampak: Kesalahan prediksi yang mengakibatkan penggunaan sumber daya yang tidak efisien, beban tambahan, dan potensi kerugian kinerja perusahaan asuransi terjadi karena calon pelaku perjalanan tidak mendapatkan klaim, meskipun telah mengajukannya. Hal ini dapat merugikan efisiensi operasional dan keuangan perusahaan.

**Tipe error 2: False Negative**

Dampak: Kesalahan prediksi yang menyebabkan menurunnya standar kualitas pelayanan asuransi untuk pelaku perjalanan. Calon pelaku perjalanan diharuskan untuk menerima klaim, meskipun sebenarnya tidak mengajukannya. Hal ini dapat merugikan reputasi perusahaan dan mempengaruhi kepercayaan pelanggan.


Dalam konteks ini, `Recall` adalah metrik evaluasi yang lebih relevan karena fokus pada mengurangi False Negatif (FN). Dengan meningkatkan `Recall`dapat membantu meminimalkan risiko kerugian reputasi dan kepercayaan pelaku perjalanan akibat kesalahan prediksi yang mengakibatkan pelaku perjalanan yang seharusnya berhak menerima klaim tidak mendapatkannya.

## **Machine Learnig**

### **Feature Engineering**

Feature engineering adalah proses manipulasi atau transformasi variabel pada dataset untuk meningkatkan performa model atau mempermudah pemahaman terhadap data. Saat Anda melakukan feature engineering, langkah-langkah umum yang dilakukan melibatkan definisi feature (X) dan target (Y), pembagian dataset, dan encoding.

**1. Define X dan Y**

```python
X = df.drop("Claim", axis=1)
Y = df["Claim"]
```

Dalam konteks penggunaan feature (X) pada dataset travel insurance, saya akan menggunakan semua kolom kecuali kolom target 'Claim'. Dengan demikian, variabel prediktor (X) akan mencakup seluruh informasi yang terdapat dalam dataset travel insurance, kecuali variabel target 'Claim'.

**Data Splitting**

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(
    X,
    Y,
    stratify=Y,
    random_state=42,
    test_size=0.2
)
```

Saya membagi data pelatihan (train) sebesar 80% dan data pengujian (test) sebesar 20%.

**Encoding Kategory**

Dengan menerapkan metode encoding ini, kita dapat  menjaga informasi yang terkandung dalam data kategorikal sambil secara efektif mengubahnya ke dalam format yang dapat diterima oleh model pembelajaran mesin.

```python
onehot_features = ['Agency Type', 'Distribution Channel']
binary_features = ['Agency', 'Product Name', 'Destination']

preprocessor = ColumnTransformer(
    transformers=[
        ('binary', ce.BinaryEncoder(cols=binary_features), binary_features),
        ('onehot', OneHotEncoder(drop='first', handle_unknown='ignore'), onehot_features)
    ],
    remainder='passthrough'
)

Xtrain_prepros = preprocessor.fit_transform(Xtrain)
Xtest_prepros = preprocessor.transform(Xtest)
```

### **Model Benchmerking**

```python
models = {
    'logreg': LogisticRegression(),
    'knn': KNeighborsClassifier(),
    'dt': DecisionTreeClassifier(),
    'rf' : RandomForestClassifier(),
    'svc' : SVC()
}
```

**Model Benchmarking : K-Fold & SMOTE**

```python
skfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
accuracy = []
recall = []

for model_name, model in models.items():

    # Training
    X_resampled, Y_resampled = SMOTE(sampling_strategy=0.5, random_state=42).fit_resample(Xtrain_prepros, Ytrain)
    
    model_pipe = Pipeline([
        ('model', model)
    ])

    model_pipe.fit(X_resampled, Y_resampled)
    
    # Evaluasi
    model_cv_accuracy = cross_val_score(model_pipe, X_resampled, Y_resampled, cv=skfold, scoring='accuracy', n_jobs=-1)
    model_cv_recall = cross_val_score(model_pipe, X_resampled, Y_resampled, cv=skfold, scoring='recall', n_jobs=-1)
    accuracy.append(model_cv_accuracy.mean() * 100)
    recall.append(model_cv_recall.mean() * 100)

# DataFrame untuk hasil k-fold
dfkFold = pd.DataFrame({
    'Model': list(models.keys()),
    'Avg Accuracy': accuracy,
    'Avg Recall': recall,
})
dfkFold.sort_values(by='Avg Recall', ascending=False, inplace=True)
```

### **Hyperparameter Tuning Menggunakan RandomizedSearchCV**

```python
skfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
rf_model = RandomForestClassifier(random_state=42)

param_distribusi = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'max_features': ['auto', 'sqrt', 'log2'],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False],
    'class_weight': ['balanced', 'balanced_subsample', None]
}
random_search = RandomizedSearchCV(
    rf_model, 
    param_distribusi, 
    n_iter=10, 
    scoring='recall', 
    cv=skfold, 
    verbose=1, 
    n_jobs=-1
)

X_resampled, Y_resampled = SMOTE(sampling_strategy=0.5, random_state=42).fit_resample(Xtrain_prepros, Ytrain)
random_search.fit(X_resampled, Y_resampled)
```
